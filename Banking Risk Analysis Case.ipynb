# Univariate Analysis

```python
bins=[0,100000,300000,float("inf")]
labels=["Low","Mid","High"]
df["Income Band"]=pd.cut(df["Estimated Income"],bins=bins,labels=labels,right=False)
```

```python
ax = df["Income Band"].value_counts().plot(kind="bar")
ax.set_xlabel("Income Band")
```

```python
categorical_cols=df[["BRId","GenderId","Amount of Credit Cards","Nationality","Fee Structure","Loyalty Classification","Occupation","Properties Owned","Risk Weighting","IAId","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts().to_frame())
```

```python
# Examine the distribution of unique categories in categorical columns

for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","IAId","Income Band"]].columns
):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

```python
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

# Numerical Analysis

```python
# Univariate Analysis and Visualization
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]
plt.figure(figsize=(20,20))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

# Heatmaps

```python
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

# Insights from the EDA

1.Bank Deposits and Checking Accounts (0.84): Very strong positive correlation, indicating that customers with higher bank deposits also tend to maintain higher checking account balances.

2.Bank Deposits and Saving Accounts (0.75): Strong positive correlation, suggesting that customers who grow their deposits also increase their savings simultaneously.

3.Business Lending and Bank Deposits (0.44): Customers with higher deposits are more likely to engage in business lending.

4.Business Lending and Bank Loans (0.42): Strong link between different forms of lending/credit.

5.Credit Card Balance and Bank Deposits (0.38), Credit Card Balance and Bank Loans (0.37): Moderate relationship between credit card usage and both deposits and loans.

6.Foreign Currency shows moderate correlations with Bank Deposits (0.41), Bank Loans (0.36), and Business Lending (0.37). This indicates that customers with foreign currency accounts are often more engaged with deposits and lending activities.

# Implications for Banking Risk Analysis

1.Checking accounts, savings accounts, and bank deposits are tightly linked, forming a natural cluster of related financial activities.

2.Business lending, bank loans, and credit card balances cluster together, highlighting a segment of customers with higher credit risk exposure.

3.Superannuation savings and estimated income are relatively independent, contributing less to cross-selling opportunities or predictive risk modeling.
